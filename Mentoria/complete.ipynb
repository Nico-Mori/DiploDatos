```python
#Cargamos un DF con todas las crypto

df_crypto = pd.DataFrame()

for coin, path in csv_paths.items():

    temp = pd.read_csv(path, index_col=0, parse_dates=True, dayfirst= True)

    cols = [f'{coin}_{c}' for c in temp.columns]
    temp.columns = cols

    df_crypto = pd.concat([df_crypto,temp], axis=1)
    
```

```python
cryptos = list(csv_paths.keys())
```

```python
for i in cryptos:
    
    first_date = df_crypto.loc[:,df_crypto.columns.str.contains(i)].first_valid_index()
    df_crypto.loc[:,df_crypto.columns.str.contains(i)].query('index >= @first_date').fillna(method='bfill', limit=36, inplace=True)

    empty_rows = df_crypto.loc[:,df_crypto.columns.str.contains(i)].query('index >= @first_date').isna().sum()[0]
    total_rows = len(df_crypto.loc[:,df_crypto.columns.str.contains(i)].query('index >= @first_date'))
    
    print('Para la crypto '+str(i)+' hay '+str(empty_rows)+' filas sin datos sobre un total de '+ str(total_rows))
```
